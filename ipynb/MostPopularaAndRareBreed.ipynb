```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from PIL import Image
from os.path import join
```

```python
RawPath = r'D:\GitWork\dog_breed\data\raw'
print("Raw path: '{}'".format(RawPath))

ImgPath = join(RawPath, 'train')
print("Image path: '{}'".format(ImgPath))


labels = pd.read_csv(join(RawPath, 'labels.csv'))
print(labels.head())

unique_breeds = labels.breed.unique()
print(); print(len(unique_breeds))

gr_labels = labels.groupby("breed").count()
gr_labels = gr_labels.rename(columns = {"id" : "count"})
gr_labels = gr_labels.sort_values("count", ascending=False)
print(gr_labels.head())
```

```python
# Show most popular breed
scottish_deerhound_id = labels.loc[labels.breed == "scottish_deerhound"].iloc[0, 0]
Image.open(join(ImgPath, scottish_deerhound_id+".jpg"))
```

```python
# Show most rare breed
print(gr_labels.tail())
```

```python
#0 row is too agressive, so i decided to take 1st :)
eskimo_dog_id = labels.loc[labels.breed == "eskimo_dog"].iloc[1, 0] 
Image.open(join(ImgPath, eskimo_dog_id+".jpg"))
```
